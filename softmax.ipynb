```python
from math import exp
```

```python
p1 = exp(1) / (exp(1) + exp(3) + exp(2))
p2 = exp(3) / (exp(1) + exp(3) + exp(2))
p3 = exp(2) / (exp(1) + exp(3) + exp(2))
print(p1, p2, p3)
print(p1 + p2 + p3)
```

```python
from numpy import exp
def softmax(vector):
 e = exp(vector)
 return e / e.sum()
data = [1, 3, 2]
result = softmax(data)
print(result)
print(sum(result))
```

```python
from scipy.special import softmax
data = [1, 3, 2]
result = softmax(data)
print(result)
print(sum(result))
```
